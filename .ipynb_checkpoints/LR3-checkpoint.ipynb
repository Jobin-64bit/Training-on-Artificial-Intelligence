```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,6,9,10,11,13,14]
y = [3,5,7,10,12,15,18, 20, 22 ,25 ]  
```

```python
plt.plot(x,y)
```

```python
A = np.polyfit(x,y,1)
```

```python
B = np.poly1d(A) 
```

```python
plt.plot(x, B(x), '--k')
plt.plot(x,y)
```
